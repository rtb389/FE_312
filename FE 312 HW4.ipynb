```python
lastdf = df[978:-1]
lastmom = newmom[972:-1]
lastbeta = newbeta[534:]
```

```python
print len(lastdf),len(lastmom),len(lastbeta)
```

```python
del lastdf['month'],lastdf['hml'],lastdf['smb'],lastdf['rf']
lastmom = lastmom[['dec1','dec10','rf']]
lastbeta = lastbeta[['dec1','dec10']]
```

```python
lastdf = lastdf.reset_index(drop=True)
lastmom =lastmom.reset_index(drop=True)
lastbeta = lastbeta.reset_index(drop=True)
```

```python
join = pd.concat([lastdf,lastmom,lastbeta],axis=1,join='inner')
```

```python
del join['month1']
```

```python
join['P1'] = -0.0092*join['me1bm1']
vals = [-1.30,0.89,-0.65,-0.42,107.66,-0.04,-1.95,-1.71,1.39,-1.50,-1.81,1.57,1.19,0.13,-4.35,0.20,0.97\
        ,-0.39,0.54,2.36,0.54,2.50,0.21,-0.36,-4.75]
for i in range(25):
    join['P1'] += vals[i]/100*join[join.columns[i+1]]
```

```python
join['P2'] = -0.7778*join['me1bm1']+1.5242*join['me1bm5']+ 1.5373*join['me5bm1']+0.2149*join['me5bm5']-1.4985*join['mkt']
```

```python
join['P3'] = -0.3253*join['me1bm1']+0.5955*join['me1bm5']+0.5578*join['me5bm1']+0.0953*join['me5bm5']-0.4911*join['mkt']\
-0.3846*join['dec1'].ix[:,0]+0.9523*join['dec10'].ix[:,0]
```

```python
join['P4'] = -0.0827*join['me1bm1']+0.1907*join['me1bm5']+0.4102*join['me5bm1']+0.0744*join['me5bm5']-0.4567*join['mkt']\
-0.1765*join['dec1'].ix[:,0]+0.4811*join['dec10'].ix[:,0]+0.5893*join['dec1'].ix[:,1]-0.0298*join['dec10'].ix[:,1]
```
