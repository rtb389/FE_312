```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'bitbootcamp/FamaDiscount.dta'
df=pd.read_stata(path)
```

```python
df.head()
```

```python
df['y1'] = -(1/float(1))*np.log(df['P1']/100)
df['y2'] = -(1/float(2))*np.log(df['P2']/100)
df['y3'] = -(1/float(3))*np.log(df['P3']/100)
df['y4'] = -(1/float(4))*np.log(df['P4']/100)
df['y5'] = -(1/float(5))*np.log(df['P5']/100)
```

```python
df.head()
```

```python
for i in [2,3,4,5]:
    df['f'+str(i)] = np.log(df['P'+str(i-1)]/100) - np.log(df['P'+str(i)]/100)
df['f1'] = - np.log(df['P1']/100)
```

```python
df.head()
```

```python
for i in [2,3,4,5]:
    df['r'+str(i)] = np.log(df['P'+str(i-1)].shift(12)/100) - np.log(df['P'+str(i)]/100)
df['r1'] = df['P1']
```

```python
df.head(15)
```

```python
from pandas.stats.api import ols
```

```python
res = []
for i in [2,3,4,5]:
    res.append(ols(y=df['y1'].shift(12*(i-1))-df['y1'],x=df['f'+str(i)]-df['y1']))
```

```python
for r in res:
    print r
```

```python
#Newey-west adjustment for serially-correlated errors
wres = []
for i in [2,3,4,5]:
    wres.append(ols(y=df['y1'].shift(12*(i-1))-df['y1'],x=df['f'+str(i)]-df['y1'],nw_lags=12*(i-1)))
```

```python
print wres
```

```python
rres = []
for i in [1,2,3,4,5]:
    rres.append(ols(y=df['r'+str(i)]-df['y1'],x=df['f'+str(i)]-df['y1'],nw_lags=12))
```

```python
print rres
```

```python
rres2 = []
for i in [1,2,3,4,5]:
    rres2.append(ols(y=df['r'+str(i)]-df['y1'],x=df[['y1','f2','f3','f4','f5']],nw_lags=12))
```

```python
print rres2
```

```python
df['CP'] = -2.4980 -0.4554*df['y1'] + 1.9560*df['f2'] -1.4094*df['f3']+0.0796*df['f4']-0.2101*df['f5']
```

```python
rres3 = []
for i in [1,2,3,4,5]:
    rres3.append(ols(y=df['r'+str(i)]-df['y1'],x=df['CP'],nw_lags=12))
```

```python
rres3
```
